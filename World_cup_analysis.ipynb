```python
import pandas as pd

# Assuming you've already uploaded the CSV file and read it into a DataFrame named 'df'

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Data Cleaning
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values:")
print(missing_values)

# Data Exploration
# Summary statistics
summary_statistics = df.describe()
print("\nSummary statistics:")
print(summary_statistics)

# Data Analysis
# Analyze team-wise performance metrics
team_performance = df.groupby('team').agg({'runs': 'sum', 'wkts': 'sum'})
print("\nTeam-wise performance metrics:")
print(team_performance)

# Visualization
import matplotlib.pyplot as plt

# Plot total runs scored by each team
team_performance.plot(kind='bar', y='runs', figsize=(10, 6), color='skyblue')
plt.title('Total Runs Scored by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot total wickets taken by each team
team_performance.plot(kind='bar', y='wkts', figsize=(10, 6), color='lightgreen')
plt.title('Total Wickets Taken by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyze individual player statistics
player_performance = df.groupby('player').agg({'runs': 'sum', 'wkts': 'sum'})
print("\nPlayer-wise performance metrics:")
print(player_performance)

# Identify leading run-scorers and wicket-takers
top_run_scorers = player_performance.nlargest(10, 'runs')
top_wicket_takers = player_performance.nlargest(10, 'wkts')
print("\nTop 10 run-scorers:")
print(top_run_scorers)
print("\nTop 10 wicket-takers:")
print(top_wicket_takers)
```

```python
# Analyze performance against different oppositions
opposition_performance = df.groupby('opposition').agg({'runs': 'sum', 'wkts': 'sum'})
print("\nPerformance against different oppositions:")
print(opposition_performance)

# Analyze performance across different grounds
ground_performance = df.groupby('ground').agg({'runs': 'sum', 'wkts': 'sum'})
print("\nPerformance across different grounds:")
print(ground_performance)
```

```python
# Study performance trends over time
temporal_analysis = df.groupby('start_date').agg({'runs': 'sum', 'wkts': 'sum'})
print("\nPerformance trends over time:")
print(temporal_analysis)

# Plot performance trends over time
temporal_analysis.plot(kind='line', y=['runs', 'wkts'], figsize=(10, 6))
plt.title('Performance Trends Over Time')
plt.xlabel('Start Date')
plt.ylabel('Total Runs/Wickets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
